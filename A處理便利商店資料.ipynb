```python
cond1 = df['分公司狀態'] == 1   #分公司狀態：01核准設立，02停業，03撤銷/廢止
cond2 = df['公司名稱'] == '統一超商股份有限公司'
cond3 = df['公司名稱'] == '全家便利商店股份有限公司'
cond4 = df['公司名稱'] == '來來超商股份有限公司'

cond5 = df['公司名稱'] == '萊爾富國際股份有限公司'
cond6 = df['公司名稱'] == '全聯實業股份有限公司'



df1 = df[cond1 & (cond2 | cond3 |cond4)]
df2 = df[cond1 & (cond5 | cond6) ]


print(len(df1))
print(df1['公司名稱'].value_counts())



print(len(df2))
print(df2['公司名稱'].value_counts())
```

# 處理後的經緯度再與原先資料庫以id或地址合併


```python
#讀取舊資料和處理完經緯度的資料
import pandas as pd
import matplotlib.pyplot as plt

from matplotlib.font_manager import *
#myfont = FontProperties(fname=r'C:\Windows\Fonts\NotoSansCJKtc-Regular.otf')
myfont = FontProperties(fname='/usr/share/fonts/opentype/noto/NotoSansCJK-Regular.ttc')
```

```python
df1 = pd.read_csv('data711FamilyOk.csv', encoding = 'utf8') 
df1 = df1.rename(columns={'Unnamed: 0': 'id'})

df2 = pd.read_csv('Address1_Finish.csv', encoding = 'utf8') 

df_result_1 = pd.merge(df1, df2, on="id")


df1 = pd.read_csv('dataHilifePxmart.csv', encoding = 'utf8') 
df1 = df1.rename(columns={'Unnamed: 0': 'id'})

df2 = pd.read_csv('Address2_Finish.csv', encoding = 'utf8') 

df_result_2 = pd.merge(df1, df2, on="id")

df = pd.concat([df_result_1, df_result_2 ])
df
```

```python
df['縣市'] = df['分公司地址'].str.slice(0, 3)
```

```python
df['縣市'].value_counts()
```

```python
#將縣市改成正確名稱
df.loc[df['縣市'].str.contains('台北縣'), '縣市'] = '新北市'
df.loc[df['縣市'].str.contains('台北市'), '縣市'] = '臺北市'
df.loc[df['縣市'].str.contains('桃園縣'), '縣市'] = '桃園市'
df.loc[df['縣市'].str.contains('彰化市'), '縣市'] = '彰化縣'
df.loc[df['縣市'].str.contains('臺南縣'), '縣市'] = '臺南市'
df.loc[df['縣市'].str.contains('苗栗市'), '縣市'] = '苗栗縣'
df.loc[df['縣市'].str.contains('臺中縣'), '縣市'] = '臺中市'
df.loc[df['縣市'].str.contains('高雄縣'), '縣市'] = '高雄市'

df['縣市'].value_counts()
```

```python
df['公司名稱'].value_counts()
```

```python
df['商店名稱'] = df['公司名稱']

df.loc[df['商店名稱'].str.contains('統一超商'), '商店名稱'] = '7-11'
df.loc[df['商店名稱'].str.contains('全家便利商店'), '商店名稱'] = '全家'
df.loc[df['商店名稱'].str.contains('萊爾富國際股份有限公司'), '商店名稱'] = '萊爾富'
df.loc[df['商店名稱'].str.contains('全聯實業股份有限公司'), '商店名稱'] = '全聯'
df.loc[df['商店名稱'].str.contains('來來超商股份有限公司'), '商店名稱'] = 'OK'

df['商店名稱'].value_counts()
```

```python
#使用pivot table
pd.pivot_table(df,index='商店名稱', columns= '縣市', values = 'id', margins=True, aggfunc=['count'])
```

```python
# groupby
gb = df.groupby('縣市')['商店名稱'].value_counts().unstack(level=1)
gb
```

```python
# plot
gb.plot.bar(stacked=True)
plt.legend(title='測試', loc='center left', bbox_to_anchor=(1, 0.5),fontproperties = myfont )
plt.show()

```

```python
#繪製五大商店的縣市分佈圖

labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 35, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]
men_std = [2, 3, 4, 1, 2]
women_std = [3, 5, 2, 3, 3]
width = 0.35       # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots()

ax.bar(labels, men_means, width, yerr=men_std, label='Men')
ax.bar(labels, women_means, width, yerr=women_std, bottom=men_means,
       label='Women')

ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.legend()

plt.show()
```
