```python
#讀取舊資料和處理完經緯度的資料
import pandas as pd
import matplotlib.pyplot as plt

from matplotlib.font_manager import *
#myfont = FontProperties(fname=r'C:\Windows\Fonts\NotoSansCJKtc-Regular.otf')
myfont = FontProperties(fname='/usr/share/fonts/opentype/noto/NotoSansCJK-Regular.ttc')
```

```python
df1 = pd.read_csv('data711FamilyOk.csv', encoding = 'utf8') 
df1 = df1.rename(columns={'Unnamed: 0': 'id'})

df2 = pd.read_csv('Address1_Finish.csv', encoding = 'utf8') 

df_result_1 = pd.merge(df1, df2, on="id")


df1 = pd.read_csv('dataHilifePxmart.csv', encoding = 'utf8') 
df1 = df1.rename(columns={'Unnamed: 0': 'id'})

df2 = pd.read_csv('Address2_Finish.csv', encoding = 'utf8') 

df_result_2 = pd.merge(df1, df2, on="id")

df = pd.concat([df_result_1, df_result_2 ])
df
```

```python
df['縣市'] = df['分公司地址'].str.slice(0, 3)
```

```python
df['縣市'].value_counts()
```

```python
#將縣市改成正確名稱
df.loc[df['縣市'].str.contains('台北縣'), '縣市'] = '新北市'
df.loc[df['縣市'].str.contains('臺北縣'), '縣市'] = '新北市'
df.loc[df['縣市'].str.contains('台北市'), '縣市'] = '臺北市'
df.loc[df['縣市'].str.contains('桃園縣'), '縣市'] = '桃園市'
df.loc[df['縣市'].str.contains('彰化市'), '縣市'] = '彰化縣'
df.loc[df['縣市'].str.contains('臺南縣'), '縣市'] = '臺南市'
df.loc[df['縣市'].str.contains('苗栗市'), '縣市'] = '苗栗縣'
df.loc[df['縣市'].str.contains('臺中縣'), '縣市'] = '臺中市'
df.loc[df['縣市'].str.contains('高雄縣'), '縣市'] = '高雄市'

df['縣市'].value_counts()
```

```python
df['公司名稱'].value_counts()
```

```python
df['商店名稱'] = df['公司名稱']

df.loc[df['商店名稱'].str.contains('統一超商'), '商店名稱'] = '7-11'
df.loc[df['商店名稱'].str.contains('全家便利商店'), '商店名稱'] = '全家'
df.loc[df['商店名稱'].str.contains('萊爾富國際股份有限公司'), '商店名稱'] = '萊爾富'
df.loc[df['商店名稱'].str.contains('全聯實業股份有限公司'), '商店名稱'] = '全聯'
df.loc[df['商店名稱'].str.contains('來來超商股份有限公司'), '商店名稱'] = 'OK'

df['商店名稱'].value_counts()
```

```python
#使用pivot table
pd.pivot_table(df,index='商店名稱', columns= '縣市', values = 'id', margins=True, aggfunc=['count'])
```

```python
# groupby
#gb = df.groupby('縣市')['商店名稱'].value_counts().unstack(level=1)

#先以縣市分類，再以各縣市的合計來排序
gb = df.groupby('縣市',sort=True)['商店名稱'].value_counts().unstack(level=1)
gb['sum'] = gb[list(gb.columns)].sum(axis=1)
gb = gb.sort_values(by='sum')
gb

```

```python
# plot


gb[['7-11','全家','萊爾富','OK','全聯']].plot.barh(stacked=True)

fig = plt.gcf()
fig.set_size_inches(8, 6)

plt.xlabel('店數', fontproperties = myfont)

plt.yticks(fontproperties = myfont  )
plt.ylabel('',)

plt.title('各縣市五大超商數量', fontproperties = myfont,size = 18)

plt.legend(loc='center left', bbox_to_anchor=(0.75, 0.5),prop = myfont )

fig.savefig('各縣市五大超商.png', dpi=200,facecolor='white', edgecolor='none', bbox_inches= 'tight')

plt.show()
```

# 以人口和面積計算人均超商數與超商密度

https://www.ris.gov.tw/info-popudata/app/awFastDownload/file/y0s6-00000.xls/y0s6/00000/

```python
df_pop = pd.read_csv('population.csv', encoding = 'utf8') 
df_pop.columns =['縣市', '人口','面積']
df_pop
```

```python
#將人口合併入gb

#gb['縣市'] = gb.index
#gb_temp = gb.reset_index(level=0, inplace=False)


df_2 = pd.merge(gb, df_pop, how="left", on=["縣市"])
#df_2['人口'].astype(int)
df_2['面積'].round(0).astype(int)
df_2['便利商店服務人數'] = df_2['人口'] / df_2['sum']
df_2['便利商店密度'] = df_2['sum'] / df_2['面積']

#df_2.sort_values(by='便利商店密度')
df_2
```

```python
x = df_2['便利商店密度']
y = df_2['便利商店服務人數']

fig, ax = plt.subplots()


fig = plt.gcf()
fig.set_size_inches(10, 10)


ax.scatter(x , y)
ax.set_xlabel('每平方公里的平均商店數量(店數/面積)',fontproperties = myfont ,size = 15)
ax.set_ylabel('便利商店平均服務人口(人口/店數)',fontproperties = myfont ,size = 15)
ax.set_title('各縣市便利商店面積密度與服務人口',fontproperties = myfont ,size = 20)
ax.grid()

for i, txt in enumerate( df_2['縣市'] ):
    ax.annotate(txt, (x[i], y[i]) ,fontproperties = myfont ,size = 12)
    
    
plt.savefig('各縣市便利商店面積密度與服務人口.png', dpi=300,facecolor='white', edgecolor='none', bbox_inches= 'tight')

plt.show()    
```

# 便利商店面積校正
### 使用 geopanda的buffer找出每個便利商店方圓200公尺的區域，取交集作為便利商店熱區
### 任何縣市的便利商店單位面積數量，最少數量為 1

```python
import geopandas
from geopandas import GeoSeries
from shapely.geometry import Polygon
from shapely.geometry import Point
import numpy as np
```

```python
city = '新竹市'
cond = df['縣市'] == city
x = df[df['Response_X'].notna() & cond ]['Response_X'].round(0).astype(int)
y = df[df['Response_Y'].notna() & cond ]['Response_Y'].round(0).astype(int)


pts = GeoSeries([Point(x, y) for x, y in zip(x, y)] )

circles = pts.buffer(200)

#合併這些circle
mp = circles.unary_union
mp
print(len(mp))
p = geopandas.GeoSeries(mp[44])
p.plot()



plt.show()
#print(city)
#print(mp.area)
```

```python
#單位圓的面積
pt = GeoSeries(Point(0, 0))
unitArea =  pt.buffer(200).area[0]



cityList = df['縣市'].value_counts(dropna=False).keys().tolist()
bufferAreaList = []

for city in cityList:
    
    cond = df['縣市'] == city
    x = df[df['Response_X'].notna() & cond ]['Response_X'].round(0).astype(int)
    y = df[df['Response_Y'].notna() & cond ]['Response_Y'].round(0).astype(int)

    store_num = len(x)
    
    pts = GeoSeries([Point(x, y) for x, y in zip(x, y)] )

    circles = pts.buffer(200)

    #合併這些circle
    mp = circles.unary_union
    mp
    #print(city)
    #print(mp.area)
    density = store_num * unitArea / mp.area
    print(density)
    bufferAreaList.append([city,density])
bufferAreaList    
```

```python
40.2*2.39/3.1
```

```python
#導入df_buffer
df_buffer = pd.DataFrame(bufferAreaList, columns =['縣市', 'buffer密度']) 

#將df_buffer併入 gd
df_3 = pd.merge(df_2, df_buffer, how="left", on=["縣市"])

df_3
```

```python

x = df_3['buffer密度']
y = df_3['便利商店服務人數']

fig, ax = plt.subplots()


fig = plt.gcf()
fig.set_size_inches(10, 10)


ax.scatter(x , y)
ax.set_xlabel('集中度(200公尺內可以找到幾家便利商店)',fontproperties = myfont ,size = 15)
ax.set_ylabel('便利商店平均服務人口(人口/店數)',fontproperties = myfont ,size = 15)
ax.set_title('在有便利商店的區域中\n集中度與服務人口',fontproperties = myfont ,size = 20)
ax.grid()

for i, txt in enumerate( df_3['縣市'] ):
    ax.annotate(txt, (x[i], y[i]) ,fontproperties = myfont ,size = 12)
    
    
plt.savefig('在有便利商店的區域中集中度與服務人口.png', dpi=300,facecolor='white', edgecolor='none', bbox_inches= 'tight')

plt.show()    
```

# 解釋用圖

```python
x1 = [2.5,5,8]
y1 = [2,6,2]

x2 = [5,6,7]
y2 = [5,6,5]


x3 = [5.9,5.2,5.1]
y3 = [5.1,5.2,5.8]

pts1 = GeoSeries([Point(x, y) for x, y in zip(x1, y1)] )
pts2 = GeoSeries([Point(x, y) for x, y in zip(x2, y2)] )
pts3 = GeoSeries([Point(x, y) for x, y in zip(x3, y3)] )

pts = [pts1,pts2,pts3]

circles1 = pts1.buffer(1.5)
circles2 = pts2.buffer(1.5)
circles3 = pts3.buffer(1.5)

#合併這些circle

p1 = geopandas.GeoSeries( circles1.unary_union)
p2 = geopandas.GeoSeries( circles2.unary_union)
p3 = geopandas.GeoSeries( circles3.unary_union)

p = [p1,p2,p3]

areaList = [circles1.unary_union.area,
            circles2.unary_union.area,
            circles3.unary_union.area]


fig, axs = plt.subplots(1,3)

fig = plt.gcf()
fig.set_size_inches(8, 3)


for idx, ax in enumerate(axs):

    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)


    p[idx].plot(ax = ax,color='gray')
    pts[idx].plot(ax =ax,color='lightblue')

    ax.set_xlim(0,10)
    ax.set_ylim(0,10)
    
    area = circles1.unary_union.area/areaList[idx]

    area = str(area)[0:3]

    ax.annotate(area, (5,8) ,fontproperties = myfont ,size = 12)


fig.suptitle('特定距離內可以遇到幾家商店', fontproperties = myfont,size = 15)
#fig.suptitle('This is a somewhat long figure title', fontsize=16)

plt.savefig('demo.png', dpi=300,facecolor='white', edgecolor='none', bbox_inches= 'tight')

plt.show()


```
