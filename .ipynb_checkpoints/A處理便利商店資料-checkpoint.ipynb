```python
import pandas as pd
```

```python
df = pd.read_csv('rawData.csv', encoding = 'utf8') 
df
```

```python
df[df['分公司狀態'] == 1]['公司名稱'].value_counts()
```

```python
cond1 = df['分公司狀態'] == 1   #分公司狀態：01核准設立，02停業，03撤銷/廢止
cond2 = df['公司名稱'] == '統一超商股份有限公司'
cond3 = df['公司名稱'] == '全家便利商店股份有限公司'
cond4 = df['公司名稱'] == '來來超商股份有限公司'

cond5 = df['公司名稱'] == '萊爾富國際股份有限公司'
cond6 = df['公司名稱'] == '全聯實業股份有限公司'



df1 = df[cond1 & (cond2 | cond3 |cond4)]
df2 = df[cond1 & (cond5 | cond6) ]


print(len(df1))
print(df1['公司名稱'].value_counts())



print(len(df2))
print(df2['公司名稱'].value_counts())
```

```python
#儲存df to csv

df1.to_csv('data711FamilyOk.csv')
df1['分公司地址'].to_csv('address711FamilyOk.csv')
df2.to_csv('dataHilifePxmart.csv')
df2['分公司地址'].to_csv('addressHilifePxmart.csv')

```
